```python
def has_no_e(word):
    if 'e' not in word:
        return True
```

        previous = c

```python
def is_triple_double(word):
    """Tests if a word contains three consecutive double letters.
    
    word: string
    returns: bool
    """
    i = 0
    count = 0
    while i < len(word)-1:
        if word[i] == word[i+1]:
            count = count + 1
            if count == 3:
                return True
            i = i + 2
        else:
            i = i + 1 - 2*count
            count = 0
    return False
```

```python
for line in open('words.txt'):
        word = line.strip()
        if is_triple_double(word):
            print(word)
```
