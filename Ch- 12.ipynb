```python
def most_frequent(string):
    dictt = {}
    for i in string:
        if i not in dictt:
            dictt[i] = 1
        elif i in dictt:
            dictt[i] += 1
    t = []
    for k,v in dictt.items():
        t.append((v,k))
    t.sort(reverse=True)
    res = []
    for a,b in t:
        res.append(b)
    return res
        
```

```python
most_frequent('aaaoosssssspp')
```

```python
def all_anagrams(filename):
    d = {}
    for line in open(filename):
        word = line.strip().lower()
        t = "".join(sorted(list(word)))
        if t not in d:
            d[t] = [word]
        elif t in d:
            d[t].append(word)
    return d        
```

```python
ana_dict = all_anagrams('words.txt')
```

```python
len(ana_dict)
```

```python
def sort_anagrams(d):
    new_dict = {}
    for i,j in d.items():
        new_dict[i] = len(j)
    t = []
    for k,v in new_dict.items():
        t.append((v,k))
    t.sort(reverse=True)
    for a,b in t:
        print(d[b])    
```

```python
for i in ana_dict.values():
    print(len(i))
    break
```

```python
sort_anagrams(ana_dict)
```

```python
def bingo(d):
    new_dict = {}
    for k,v in d.items():
        if len(k) == 8 :
            new_dict[k] = v
    return new_dict
```
